```python
import pandas as pd
```

```python
data = {
    'Nome': ['Pikachu', 'Charmander', 'Bulbasaur', 'Squirtle', 'Jigglypuff', 'Gengar', 'Eevee', 'Snorlax', 'Mewtwo', 'Lucario'],
    'Tipo': ['Elétrico', 'Fogo', 'Grama/Veneno', 'Água', 'Normal/Fada', 'Fantasma/Veneno', 'Normal', 'Normal', 'Psíquico', 'Lutador/Aço'],
    'HP': [35, 39, 45, 44, 115, 60, 55, 160, 106, 70],
    'Ataque': [55, 52, 49, 48, 45, 65, 55, 110, 110, 110],
    'Defesa': [40, 43, 49, 65, 20, 60, 50, 65, 90, 70],
    'Velocidade': [90, 65, 45, 43, 20, 110, 55, 30, 130, 90]
}

df = pd.DataFrame(data)
print(df)
```

```python
df['Total'] = df['HP'] + df['Ataque'] + df['Defesa'] + df['Velocidade']
df
```

```python
df[df['Velocidade'] > 80]
```

```python
df.sort_values('Total', ascending=True)
```

```python
media_pokemon = df.groupby('Ataque')[['Defesa', 'Velocidade', 'HP',]].mean()
media_pokemon
```

```python

 devio_padrao = df.groupby('Ataque')[['Defesa', 'Velocidade', 'HP',]].std()
 devio_padrao
```

```python
import matplotlib.pyplot as plt
```

```python
df.boxplot(column='HP', by='Tipo', grid=False, rot=90)
plt.title('Distribuição de HP por Tipo de Pokémon')
plt.suptitle('')
plt.xlabel('Tipo')
plt.ylabel('HP')
plt.show()
```

```python
df.boxplot(column='Ataque', by='Tipo', grid=False, rot=90)
plt.title('Distribuição de Ataque por Tipo de Pokémon')
plt.suptitle('')
plt.xlabel('Tipo')
plt.ylabel('Ataque')
plt.show()
```

```python
df.boxplot(column='Defesa', by='Tipo', grid=False, rot=90)
plt.title('Distribuição de Defesa por Tipo de Pokémon')
plt.suptitle('')
plt.xlabel('Tipo')
plt.ylabel('Defesa')
plt.show()
```

```python
df.boxplot(column='Velocidade', by='Tipo', grid=False, rot=90)
plt.title('Distribuição de Velocidade por Tipo de Pokémon')
plt.suptitle('')
plt.xlabel('Tipo')
plt.ylabel('Velocidade')
plt.show()
```
